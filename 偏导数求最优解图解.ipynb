````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# ËÆæÁΩÆ‰∏≠ÊñáÊòæÁ§∫
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei']
plt.rcParams['axes.unicode_minus'] = False

print("‚úÖ Â∫ìÂØºÂÖ•ÊàêÂäüÔºÅ")
````

## üìê Á¨¨‰∏ÄÊ≠•ÔºöÂõûÈ°æÈ´ò‰∏≠Áü•ËØÜ - ‰∏ÄÂÖÉÂáΩÊï∞Ê±ÇÊûÅÂÄº

ÂáΩÊï∞Ôºöf(x) = x¬≤ - 4x + 5

````python
# ‰∏ÄÂÖÉÂáΩÊï∞
x = np.linspace(-1, 5, 100)
y = x**2 - 4*x + 5

# ÂØºÊï∞‰∏∫0ÁöÑÁÇπ
x_min = 2
y_min = x_min**2 - 4*x_min + 5

# ÁªòÂõæ
plt.figure(figsize=(12, 5))

# Â∑¶ÂõæÔºöÂáΩÊï∞Êõ≤Á∫ø
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = x¬≤ - 4x + 5')
plt.plot(x_min, y_min, 'ro', markersize=12, label=f'ÊúÄÂ∞èÂÄºÁÇπ ({x_min}, {y_min})')
plt.axhline(y=y_min, color='r', linestyle='--', alpha=0.3)
plt.axvline(x=x_min, color='r', linestyle='--', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('‰∏ÄÂÖÉÂáΩÊï∞ÁöÑÊúÄÂ∞èÂÄº', fontsize=14, fontweight='bold')
plt.legend()

# Âè≥ÂõæÔºöÂØºÊï∞
plt.subplot(1, 2, 2)
x_deriv = np.linspace(-1, 5, 100)
y_deriv = 2*x_deriv - 4
plt.plot(x_deriv, y_deriv, 'g-', linewidth=2, label="f'(x) = 2x - 4")
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.axvline(x=x_min, color='r', linestyle='--', alpha=0.5)
plt.plot(x_min, 0, 'ro', markersize=12, label=f"f'({x_min}) = 0")
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=12)
plt.ylabel("f'(x)", fontsize=12)
plt.title('ÂØºÊï∞‰∏∫0ÁöÑÁÇπ = ÊûÅÂÄºÁÇπ', fontsize=14, fontweight='bold')
plt.legend()

plt.tight_layout()
plt.show()

print("\nüí° ÂÖ≥ÈîÆÔºöÂØºÊï∞‰∏∫0ÁöÑÂú∞ÊñπÔºåÂàáÁ∫øÊòØÊ∞¥Âπ≥ÁöÑÔºåËøôÈáåÂ∞±ÊòØÊûÅÂÄºÁÇπÔºÅ")
````

## üèîÔ∏è Á¨¨‰∫åÊ≠•Ôºö‰∫åÂÖÉÂáΩÊï∞ - ÂÉè‰∏ÄÂ∫ßÂ±±

ÂáΩÊï∞Ôºöf(x, y) = x¬≤ + y¬≤ - 4x - 6y + 13

ËøôÊòØ‰∏Ä‰∏™Á¢óÂΩ¢Êõ≤Èù¢ÔºåÊúÄ‰ΩéÁÇπÂú®Âì™ÈáåÔºü

````python
# ÂàõÂª∫ÁΩëÊ†º
x = np.linspace(-2, 6, 100)
y = np.linspace(-2, 8, 100)
X, Y = np.meshgrid(x, y)

# ‰∫åÂÖÉÂáΩÊï∞
Z = X**2 + Y**2 - 4*X - 6*Y + 13

# Ê±ÇÂÅèÂØºÊâæÊúÄÂ∞èÂÄºÁÇπ
# ‚àÇf/‚àÇx = 2x - 4 = 0 ‚Üí x = 2
# ‚àÇf/‚àÇy = 2y - 6 = 0 ‚Üí y = 3
x_min, y_min = 2, 3
z_min = x_min**2 + y_min**2 - 4*x_min - 6*y_min + 13

# ÁªòÂà∂3DÂõæ
fig = plt.figure(figsize=(16, 6))

# 3DÊõ≤Èù¢Âõæ
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap=cm.coolwarm, alpha=0.8)
ax1.scatter([x_min], [y_min], [z_min], color='red', s=200, marker='*', 
            label=f'ÊúÄÂ∞èÂÄºÁÇπ ({x_min}, {y_min}, {z_min})')
ax1.set_xlabel('x', fontsize=11)
ax1.set_ylabel('y', fontsize=11)
ax1.set_zlabel('f(x,y)', fontsize=11)
ax1.set_title('‰∫åÂÖÉÂáΩÊï∞ - Á¢óÂΩ¢Êõ≤Èù¢', fontsize=14, fontweight='bold')
ax1.legend()

# Á≠âÈ´òÁ∫øÂõæ
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, levels=20, cmap=cm.coolwarm)
ax2.clabel(contour, inline=True, fontsize=8)
ax2.plot(x_min, y_min, 'r*', markersize=20, label=f'ÊúÄÂ∞èÂÄºÁÇπ ({x_min}, {y_min})')
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('Á≠âÈ´òÁ∫øÂõæÔºà‰øØËßÜÂõæÔºâ', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()

print(f"\n‚úÖ ÊúÄÂ∞èÂÄºÁÇπÔºöx={x_min}, y={y_min}, f(x,y)={z_min}")
print("üí° Á∫¢ÊòüÊ†áËÆ∞ÁöÑÂú∞ÊñπÂ∞±ÊòØÁ¢óÂ∫ï - ÊúÄ‰ΩéÁÇπÔºÅ")
````

## üìä Á¨¨ÂõõÊ≠•ÔºöÊúÄÂ∞è‰∫å‰πòÊ≥ïÂÆû‰æã

Áé∞Âú®Áî®ÁúüÂÆûÊï∞ÊçÆÊºîÁ§∫ÂÆåÊï¥ÁöÑÂÅèÂØºÊï∞Ê±ÇËß£ËøáÁ®ã

````python
# ÁîüÊàêÁ§∫‰æãÊï∞ÊçÆ
np.random.seed(42)
X_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.1, 3.9, 6.2, 7.8, 10.1])  # ÁúüÂÆûÂÄºÔºàÂ∏¶Âô™Â£∞Ôºâ

# ÁªòÂà∂Êï∞ÊçÆÁÇπ
plt.figure(figsize=(10, 6))
plt.scatter(X_data, y_data, color='red', s=100, zorder=3, label='ÁúüÂÆûÊï∞ÊçÆÁÇπ')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('ÂéüÂßãÊï∞ÊçÆÁÇπ', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f"Êï∞ÊçÆÁÇπÔºö")
for i in range(len(X_data)):
    print(f"  ÁÇπ{i+1}: (x={X_data[i]}, y={y_data[i]})")
````

### üßÆ ÊâãÂä®Êé®ÂØºÊúÄÂ∞è‰∫å‰πòÊ≥ï

ÁõÆÊ†áÔºöÊâæÂà∞ÊúÄ‰ºòÁöÑ w Âíå bÔºå‰ΩøÂæó y = wx + b ÊúÄÊé•ËøëÊâÄÊúâÊï∞ÊçÆÁÇπ

````python
print("="*80)
print("üìê ÊúÄÂ∞è‰∫å‰πòÊ≥ïÔºöÂÆåÊï¥Êé®ÂØºËøáÁ®ã")
print("="*80)

n = len(X_data)
sum_x = np.sum(X_data)
sum_y = np.sum(y_data)
sum_xy = np.sum(X_data * y_data)
sum_x2 = np.sum(X_data ** 2)
x_mean = np.mean(X_data)
y_mean = np.mean(y_data)

print("\n„ÄêÊï∞ÊçÆÁªüËÆ°„Äë")
print(f"Ê†∑Êú¨Êï∞Èáè n = {n}")
print(f"Œ£x = {sum_x}")
print(f"Œ£y = {sum_y:.2f}")
print(f"Œ£xy = {sum_xy:.2f}")
print(f"Œ£x¬≤ = {sum_x2}")
print(f"xÃÑ = {x_mean}")
print(f"»≥ = {y_mean:.2f}")

print("\n„ÄêÊ≠•È™§1„ÄëÂª∫Á´ãÊçüÂ§±ÂáΩÊï∞")
print("-" * 80)
print("J(w, b) = Œ£(y·µ¢ - (wx·µ¢ + b))¬≤")

print("\n„ÄêÊ≠•È™§2„ÄëÂØπ w Ê±ÇÂÅèÂØº")
print("-" * 80)
print("‚àÇJ/‚àÇw = Œ£ 2(y·µ¢ - wx·µ¢ - b)¬∑(-x·µ¢)")
print("      = -2Œ£ x·µ¢(y·µ¢ - wx·µ¢ - b)")
print("      = -2(Œ£x·µ¢y·µ¢ - wŒ£x·µ¢¬≤ - bŒ£x·µ¢)")
print("\n‰ª§ ‚àÇJ/‚àÇw = 0:")
print("Œ£x·µ¢y·µ¢ - wŒ£x·µ¢¬≤ - bŒ£x·µ¢ = 0  ... ÊñπÁ®ã‚ë†")

print("\n„ÄêÊ≠•È™§3„ÄëÂØπ b Ê±ÇÂÅèÂØº")
print("-" * 80)
print("‚àÇJ/‚àÇb = Œ£ 2(y·µ¢ - wx·µ¢ - b)¬∑(-1)")
print("      = -2Œ£(y·µ¢ - wx·µ¢ - b)")
print("      = -2(Œ£y·µ¢ - wŒ£x·µ¢ - nb)")
print("\n‰ª§ ‚àÇJ/‚àÇb = 0:")
print("Œ£y·µ¢ - wŒ£x·µ¢ - nb = 0")
print("b = (Œ£y·µ¢ - wŒ£x·µ¢) / n")
print("b = »≥ - w¬∑xÃÑ  ... ÊñπÁ®ã‚ë°")

print("\n„ÄêÊ≠•È™§4„ÄëÊ±ÇËß£ w")
print("-" * 80)
print("Â∞ÜÊñπÁ®ã‚ë°‰ª£ÂÖ•ÊñπÁ®ã‚ë†:")
print("Œ£x·µ¢y·µ¢ - wŒ£x·µ¢¬≤ - (»≥ - w¬∑xÃÑ)¬∑Œ£x·µ¢ = 0")
print("Œ£x·µ¢y·µ¢ - wŒ£x·µ¢¬≤ - »≥¬∑Œ£x·µ¢ + w¬∑xÃÑ¬∑Œ£x·µ¢ = 0")
print("w(Œ£x·µ¢¬≤ - xÃÑ¬∑Œ£x·µ¢) = »≥¬∑Œ£x·µ¢ - Œ£x·µ¢y·µ¢")
print("\nÊ≥®ÊÑèÔºöŒ£x·µ¢ = n¬∑xÃÑÔºåÊâÄ‰ª• xÃÑ¬∑Œ£x·µ¢ = n¬∑xÃÑ¬≤")
print("\nw = (Œ£x·µ¢y·µ¢ - n¬∑xÃÑ¬∑»≥) / (Œ£x·µ¢¬≤ - n¬∑xÃÑ¬≤)")

# ËÆ°ÁÆó w Âíå b
w = (sum_xy - n * x_mean * y_mean) / (sum_x2 - n * x_mean**2)
b = y_mean - w * x_mean

print("\n„ÄêÊ≠•È™§5„Äë‰ª£ÂÖ•Êï∞ÂÄºËÆ°ÁÆó")
print("-" * 80)
print(f"w = ({sum_xy:.2f} - {n}√ó{x_mean}√ó{y_mean:.2f}) / ({sum_x2} - {n}√ó{x_mean}¬≤)")
print(f"  = ({sum_xy:.2f} - {n * x_mean * y_mean:.2f}) / ({sum_x2} - {n * x_mean**2:.2f})")
print(f"  = {sum_xy - n * x_mean * y_mean:.2f} / {sum_x2 - n * x_mean**2:.2f}")
print(f"  = {w:.4f}")

print(f"\nb = »≥ - w¬∑xÃÑ")
print(f"  = {y_mean:.2f} - {w:.4f}√ó{x_mean}")
print(f"  = {b:.4f}")

print("\n" + "="*80)
print(f"‚úÖ ÊúÄ‰ºòËß£Ôºöw = {w:.4f}, b = {b:.4f}")
print(f"‚úÖ ÊãüÂêàÁõ¥Á∫øÔºöy = {w:.4f}x + {b:.4f}")
print("="*80)
````

````python
# ÂèØËßÜÂåñÊãüÂêàÁªìÊûú
x_line = np.linspace(0, 6, 100)
y_line = w * x_line + b

plt.figure(figsize=(10, 6))
plt.scatter(X_data, y_data, color='red', s=100, zorder=3, label='ÁúüÂÆûÊï∞ÊçÆÁÇπ')
plt.plot(x_line, y_line, 'b-', linewidth=2, label=f'ÊãüÂêàÁõ¥Á∫ø: y = {w:.2f}x + {b:.2f}')

# ÁªòÂà∂ËØØÂ∑ÆÁ∫ø
for i in range(len(X_data)):
    y_pred_i = w * X_data[i] + b
    plt.plot([X_data[i], X_data[i]], [y_data[i], y_pred_i], 'g--', alpha=0.5)

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('ÊúÄÂ∞è‰∫å‰πòÊ≥ïÊãüÂêàÁªìÊûú', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# ËÆ°ÁÆóÊçüÂ§±
y_pred = w * X_data + b
loss = np.sum((y_data - y_pred)**2)
print(f"\nÊÄªÊçüÂ§±ÔºàËØØÂ∑ÆÂπ≥ÊñπÂíåÔºâ: {loss:.4f}")
print("üí° ÁªøËâ≤ËôöÁ∫øÊòØÊØè‰∏™ÁÇπÁöÑËØØÂ∑ÆÔºåÊàë‰ª¨ÈÄöËøáÊ±ÇÂÅèÂØºËÆ©Ëøô‰∫õËØØÂ∑ÆÁöÑÂπ≥ÊñπÂíåÊúÄÂ∞èÔºÅ")
````
