````python
"""
æ–¹æ¡ˆä¸€ï¼šç®€å•ç‰ˆ Agentic RAG æ™ºèƒ½å®¢æœ

æ ¸å¿ƒåŠŸèƒ½ï¼š
1. çŸ¥è¯†åº“æ£€ç´¢ï¼ˆRAGï¼‰
2. åŸºç¡€å·¥å…·è°ƒç”¨ï¼ˆè®¢å•æŸ¥è¯¢ï¼‰
3. ç®€å•çš„æ„å›¾è¯†åˆ«
4. å¯¹è¯è®°å¿†
"""

import json
import numpy as np
from typing import List, Dict, Any
from datetime import datetime

# ========== 1. æ¨¡æ‹Ÿçš„å·¥å…·å‡½æ•° ==========

def query_order(order_id: str) -> Dict[str, Any]:
    """æ¨¡æ‹Ÿè®¢å•æŸ¥è¯¢ç³»ç»Ÿ"""
    # å®é™…åº”è¯¥è¿æ¥çœŸå®æ•°æ®åº“
    mock_orders = {
        "ORDER123456": {
            "status": "å·²å‘è´§",
            "product": "iPhone 15",
            "delivery_date": "2024-11-28",
            "tracking_number": "SF123456789"
        },
        "ORDER789012": {
            "status": "å¤„ç†ä¸­",
            "product": "MacBook Pro",
            "delivery_date": "é¢„è®¡12æœˆ5æ—¥",
            "tracking_number": "æš‚æ— "
        }
    }
    
    if order_id in mock_orders:
        return {
            "success": True,
            "data": mock_orders[order_id]
        }
    else:
        return {
            "success": False,
            "message": "è®¢å•ä¸å­˜åœ¨"
        }


def check_product_info(product_name: str) -> Dict[str, Any]:
    """æ¨¡æ‹Ÿäº§å“ä¿¡æ¯æŸ¥è¯¢"""
    mock_products = {
        "iPhone 15": {
            "price": 5999,
            "stock": "æœ‰è´§",
            "description": "å…¨æ–° A17 Pro èŠ¯ç‰‡ï¼Œé’›é‡‘å±æœºèº«"
        },
        "MacBook Pro": {
            "price": 14999,
            "stock": "æœ‰è´§",
            "description": "M3 èŠ¯ç‰‡ï¼Œ18å°æ—¶ç»­èˆª"
        }
    }
    
    for key in mock_products:
        if product_name.lower() in key.lower():
            return {
                "success": True,
                "data": mock_products[key]
            }
    
    return {
        "success": False,
        "message": "æœªæ‰¾åˆ°è¯¥äº§å“"
    }


# ========== 2. ç®€å•çš„çŸ¥è¯†åº“ï¼ˆRAGï¼‰==========

class SimpleKnowledgeBase:
    """ç®€å•çš„çŸ¥è¯†åº“å®ç°"""
    
    def __init__(self):
        self.knowledge = [
            {
                "question": "å¦‚ä½•é€€è´§",
                "answer": "é€€è´§æµç¨‹ï¼š1) åœ¨è®¢å•è¯¦æƒ…ç‚¹å‡»é€€è´§ 2) é€‰æ‹©é€€è´§åŸå›  3) æäº¤ç”³è¯· 4) ç­‰å¾…å®¡æ ¸"
            },
            {
                "question": "è¿è´¹æ”¿ç­–",
                "answer": "æ»¡99å…ƒåŒ…é‚®ï¼Œä¸æ»¡99å…ƒè¿è´¹10å…ƒã€‚åè¿œåœ°åŒºè¿è´¹å¦è®¡ã€‚"
            },
            {
                "question": "ä¿ä¿®æœŸå¤šä¹…",
                "answer": "æ‰€æœ‰äº§å“äº«å—1å¹´è´¨ä¿ï¼Œéƒ¨åˆ†äº§å“å¯å»¶ä¿è‡³3å¹´ã€‚"
            },
            {
                "question": "æ”¯ä»˜æ–¹å¼",
                "answer": "æ”¯æŒå¾®ä¿¡æ”¯ä»˜ã€æ”¯ä»˜å®ã€ä¿¡ç”¨å¡ã€èŠ±å‘—åˆ†æœŸç­‰å¤šç§æ”¯ä»˜æ–¹å¼ã€‚"
            }
        ]
    
    def search(self, query: str, top_k: int = 2) -> List[Dict]:
        """ç®€å•çš„å…³é”®è¯åŒ¹é…æ£€ç´¢"""
        results = []
        for item in self.knowledge:
            # ç®€å•çš„å…³é”®è¯åŒ¹é…
            if any(word in item["question"] or word in item["answer"] 
                   for word in query):
                results.append(item)
        
        return results[:top_k]


# ========== 3. Agent æ™ºèƒ½å†³ç­– ==========

class SimpleCustomerServiceAgent:
    """ç®€å•ç‰ˆæ™ºèƒ½å®¢æœ Agent"""
    
    def __init__(self, api_key: str = None):
        self.api_key = api_key
        self.knowledge_base = SimpleKnowledgeBase()
        self.conversation_history = []
        
        # å®šä¹‰å·¥å…·
        self.tools = {
            "query_order": query_order,
            "check_product_info": check_product_info
        }
    
    def classify_intent(self, message: str) -> str:
        """ç®€å•çš„æ„å›¾è¯†åˆ«"""
        message_lower = message.lower()
        
        # è®¢å•æŸ¥è¯¢
        if any(word in message_lower for word in ["è®¢å•", "æŸ¥è¯¢è®¢å•", "order"]):
            return "query_order"
        
        # äº§å“å’¨è¯¢
        if any(word in message_lower for word in ["äº§å“", "ä»·æ ¼", "åº“å­˜", "iphone", "macbook"]):
            return "product_inquiry"
        
        # é€€æ¢è´§
        if any(word in message_lower for word in ["é€€è´§", "é€€æ¬¾", "æ¢è´§"]):
            return "return_refund"
        
        # æŠ•è¯‰
        if any(word in message_lower for word in ["æŠ•è¯‰", "ä¸æ»¡æ„", "å·®è¯„"]):
            return "complaint"
        
        # é»˜è®¤ï¼šçŸ¥è¯†åº“æ£€ç´¢
        return "knowledge_base"
    
    def extract_order_id(self, message: str) -> str:
        """æå–è®¢å•å·"""
        import re
        match = re.search(r'ORDER\d+', message.upper())
        if match:
            return match.group()
        return None
    
    def process(self, user_message: str) -> str:
        """å¤„ç†ç”¨æˆ·æ¶ˆæ¯"""
        
        # è®°å½•å¯¹è¯
        self.conversation_history.append({
            "role": "user",
            "content": user_message,
            "timestamp": datetime.now().isoformat()
        })
        
        # 1. è¯†åˆ«æ„å›¾
        intent = self.classify_intent(user_message)
        print(f"ğŸ¤” è¯†åˆ«æ„å›¾: {intent}")
        
        # 2. æ ¹æ®æ„å›¾æ‰§è¡Œæ“ä½œ
        response = ""
        
        if intent == "query_order":
            # æå–è®¢å•å·
            order_id = self.extract_order_id(user_message)
            
            if order_id:
                # æŸ¥è¯¢è®¢å•
                result = self.tools["query_order"](order_id)
                
                if result["success"]:
                    data = result["data"]
                    response = f"æ‚¨çš„è®¢å• {order_id} çŠ¶æ€å¦‚ä¸‹ï¼š\n"
                    response += f"â€¢ å•†å“ï¼š{data['product']}\n"
                    response += f"â€¢ çŠ¶æ€ï¼š{data['status']}\n"
                    response += f"â€¢ é¢„è®¡é€è¾¾ï¼š{data['delivery_date']}\n"
                    if data['tracking_number'] != "æš‚æ— ":
                        response += f"â€¢ ç‰©æµå•å·ï¼š{data['tracking_number']}"
                else:
                    response = result["message"]
            else:
                response = "è¯·æä¾›æ‚¨çš„è®¢å•å·ï¼ˆæ ¼å¼ï¼šORDER + æ•°å­—ï¼‰"
        
        elif intent == "product_inquiry":
            # æ£€ç´¢äº§å“ä¿¡æ¯
            result = self.tools["check_product_info"](user_message)
            
            if result["success"]:
                data = result["data"]
                response = f"äº§å“ä¿¡æ¯ï¼š\n"
                response += f"â€¢ ä»·æ ¼ï¼šÂ¥{data['price']}\n"
                response += f"â€¢ åº“å­˜ï¼š{data['stock']}\n"
                response += f"â€¢ æè¿°ï¼š{data['description']}"
            else:
                # ä»çŸ¥è¯†åº“æ£€ç´¢
                kb_results = self.knowledge_base.search(user_message)
                if kb_results:
                    response = kb_results[0]["answer"]
                else:
                    response = "æŠ±æ­‰ï¼Œæœªæ‰¾åˆ°ç›¸å…³äº§å“ä¿¡æ¯ã€‚æ‚¨å¯ä»¥æè¿°æ›´å¤šç»†èŠ‚å—ï¼Ÿ"
        
        elif intent == "complaint":
            response = "éå¸¸æŠ±æ­‰ç»™æ‚¨å¸¦æ¥ä¸ä¾¿ã€‚æ‚¨çš„æŠ•è¯‰å¾ˆé‡è¦ï¼Œæˆ‘å°†ä¸ºæ‚¨è½¬æ¥äººå·¥å®¢æœå¤„ç†ã€‚è¯·ç¨ç­‰..."
        
        else:
            # é»˜è®¤ï¼šçŸ¥è¯†åº“æ£€ç´¢
            kb_results = self.knowledge_base.search(user_message)
            
            if kb_results:
                response = kb_results[0]["answer"]
            else:
                response = "æŠ±æ­‰ï¼Œæˆ‘æ²¡æœ‰æ‰¾åˆ°ç›¸å…³ä¿¡æ¯ã€‚è¯·é—®æ‚¨è¿˜æœ‰ä»€ä¹ˆéœ€è¦å¸®åŠ©çš„å—ï¼Ÿ"
        
        # è®°å½•å›å¤
        self.conversation_history.append({
            "role": "assistant",
            "content": response,
            "timestamp": datetime.now().isoformat()
        })
        
        return response
    
    def get_conversation_history(self) -> List[Dict]:
        """è·å–å¯¹è¯å†å²"""
        return self.conversation_history


# ========== 4. æµ‹è¯•è¿è¡Œ ==========

print("=" * 60)
print("ğŸ¤– ç®€å•ç‰ˆ Agentic RAG æ™ºèƒ½å®¢æœ")
print("=" * 60)

# åˆ›å»ºå®¢æœ Agent
agent = SimpleCustomerServiceAgent()

# æ¨¡æ‹Ÿå¯¹è¯
test_conversations = [
    "ä½ å¥½ï¼Œæˆ‘æƒ³æŸ¥è¯¢æˆ‘çš„è®¢å•",
    "ORDER123456",
    "iPhone 15 å¤šå°‘é’±ï¼Ÿ",
    "å¦‚ä½•é€€è´§ï¼Ÿ",
    "ä½ ä»¬çš„å®¢æœå¤ªå·®äº†ï¼Œæˆ‘è¦æŠ•è¯‰ï¼"
]

for i, message in enumerate(test_conversations, 1):
    print(f"\n{'='*60}")
    print(f"ğŸ’¬ ç”¨æˆ· ({i}): {message}")
    print(f"{'='*60}")
    
    response = agent.process(message)
    
    print(f"\nğŸ¤– å®¢æœ: {response}")

# æ˜¾ç¤ºå®Œæ•´å¯¹è¯å†å²
print("\n\n" + "=" * 60)
print("ğŸ“œ å®Œæ•´å¯¹è¯å†å²")
print("=" * 60)

for i, msg in enumerate(agent.get_conversation_history(), 1):
    role = "ç”¨æˆ·" if msg["role"] == "user" else "å®¢æœ"
    print(f"\n{i}. [{role}] {msg['timestamp']}")
    print(f"   {msg['content']}")
````

````python
"""
æ–¹æ¡ˆäºŒï¼šç”Ÿäº§çº§ Agentic RAG æ™ºèƒ½å®¢æœï¼ˆLangGraph ç‰ˆï¼‰

æ ¸å¿ƒå‡çº§ï¼š
1. ä½¿ç”¨çœŸå® LLM è¿›è¡Œæ„å›¾ç†è§£å’Œå†³ç­–
2. å‘é‡åŒ–çŸ¥è¯†åº“ï¼ˆè¯­ä¹‰æ£€ç´¢ï¼‰
3. ä½¿ç”¨ LangGraph å®ç°å¤æ‚å·¥ä½œæµ
4. Function Callingï¼ˆå·¥å…·è°ƒç”¨ï¼‰
"""

from typing import TypedDict, Annotated, List, Dict
from langgraph.graph import StateGraph, END
import operator

# ========== 1. å®šä¹‰ Agent çŠ¶æ€ ==========

class AgentState(TypedDict):
    """Agent çš„çŠ¶æ€å®šä¹‰"""
    messages: Annotated[List[Dict], operator.add]  # å¯¹è¯å†å²
    user_query: str  # å½“å‰ç”¨æˆ·é—®é¢˜
    intent: str  # è¯†åˆ«çš„æ„å›¾
    need_tool: bool  # æ˜¯å¦éœ€è¦è°ƒç”¨å·¥å…·
    tool_name: str  # å·¥å…·åç§°
    tool_result: str  # å·¥å…·ç»“æœ
    final_answer: str  # æœ€ç»ˆå›ç­”
    iteration: int  # è¿­ä»£æ¬¡æ•°


# ========== 2. å·¥å…·å®šä¹‰ï¼ˆä¸æ–¹æ¡ˆä¸€ç›¸åŒï¼‰==========

# å¤ç”¨æ–¹æ¡ˆä¸€çš„å·¥å…·å‡½æ•°
# query_order, check_product_info

tools_description = """
å¯ç”¨å·¥å…·åˆ—è¡¨ï¼š

1. query_order
   - åŠŸèƒ½ï¼šæŸ¥è¯¢è®¢å•çŠ¶æ€
   - å‚æ•°ï¼šorder_id (è®¢å•å·ï¼Œæ ¼å¼ï¼šORDER + æ•°å­—)
   - è¿”å›ï¼šè®¢å•çŠ¶æ€ã€å•†å“ã€é…é€ä¿¡æ¯

2. check_product_info
   - åŠŸèƒ½ï¼šæŸ¥è¯¢äº§å“ä¿¡æ¯
   - å‚æ•°ï¼šproduct_name (äº§å“åç§°)
   - è¿”å›ï¼šä»·æ ¼ã€åº“å­˜ã€äº§å“æè¿°

3. transfer_human
   - åŠŸèƒ½ï¼šè½¬æ¥äººå·¥å®¢æœ
   - å‚æ•°ï¼šreason (è½¬æ¥åŸå› )
   - è¿”å›ï¼šè½¬æ¥æˆåŠŸæ¶ˆæ¯
"""


# ========== 3. Agent èŠ‚ç‚¹å‡½æ•° ==========

def intent_classifier(state: AgentState) -> AgentState:
    """èŠ‚ç‚¹1ï¼šæ„å›¾åˆ†ç±»"""
    user_query = state["user_query"]
    
    # åœ¨å®é™…åº”ç”¨ä¸­ï¼Œè¿™é‡Œä¼šè°ƒç”¨ LLM
    # ä½¿ç”¨ prompt è®© LLM åˆ†ææ„å›¾
    
    prompt = f"""
    ç”¨æˆ·é—®é¢˜ï¼š{user_query}
    
    è¯·åˆ†æç”¨æˆ·æ„å›¾ï¼Œä»ä»¥ä¸‹é€‰é¡¹ä¸­é€‰æ‹©ï¼š
    1. query_order - æŸ¥è¯¢è®¢å•
    2. product_inquiry - äº§å“å’¨è¯¢
    3. return_refund - é€€æ¢è´§é—®é¢˜
    4. complaint - æŠ•è¯‰
    5. general_question - ä¸€èˆ¬é—®é¢˜
    
    åªè¿”å›æ„å›¾ç±»å‹ã€‚
    """
    
    # ç®€åŒ–ç‰ˆï¼šä½¿ç”¨å…³é”®è¯ï¼ˆå®é™…åº”è°ƒç”¨ LLM APIï¼‰
    user_lower = user_query.lower()
    
    if "è®¢å•" in user_lower or "order" in user_lower:
        intent = "query_order"
    elif "äº§å“" in user_lower or "ä»·æ ¼" in user_lower:
        intent = "product_inquiry"
    elif "æŠ•è¯‰" in user_lower:
        intent = "complaint"
    else:
        intent = "general_question"
    
    print(f"ğŸ¤” æ„å›¾åˆ†ç±»: {intent}")
    
    return {
        **state,
        "intent": intent
    }


def tool_selector(state: AgentState) -> AgentState:
    """èŠ‚ç‚¹2ï¼šé€‰æ‹©å·¥å…·"""
    intent = state["intent"]
    
    # æ ¹æ®æ„å›¾å†³å®šæ˜¯å¦éœ€è¦å·¥å…·
    if intent == "query_order":
        need_tool = True
        tool_name = "query_order"
    elif intent == "product_inquiry":
        need_tool = True
        tool_name = "check_product_info"
    elif intent == "complaint":
        need_tool = True
        tool_name = "transfer_human"
    else:
        need_tool = False
        tool_name = ""
    
    print(f"ğŸ”§ å·¥å…·é€‰æ‹©: {tool_name if need_tool else 'æ— éœ€å·¥å…·'}")
    
    return {
        **state,
        "need_tool": need_tool,
        "tool_name": tool_name
    }


def tool_executor(state: AgentState) -> AgentState:
    """èŠ‚ç‚¹3ï¼šæ‰§è¡Œå·¥å…·"""
    tool_name = state["tool_name"]
    user_query = state["user_query"]
    
    tool_result = ""
    
    if tool_name == "query_order":
        # æå–è®¢å•å·
        import re
        match = re.search(r'ORDER\d+', user_query.upper())
        
        if match:
            order_id = match.group()
            result = query_order(order_id)
            
            if result["success"]:
                data = result["data"]
                tool_result = f"è®¢å•ä¿¡æ¯ï¼š{data}"
            else:
                tool_result = result["message"]
        else:
            tool_result = "æœªæ‰¾åˆ°è®¢å•å·ï¼Œè¯·æä¾›è®¢å•å·"
    
    elif tool_name == "check_product_info":
        result = check_product_info(user_query)
        
        if result["success"]:
            tool_result = f"äº§å“ä¿¡æ¯ï¼š{result['data']}"
        else:
            tool_result = result["message"]
    
    elif tool_name == "transfer_human":
        tool_result = "å·²ä¸ºæ‚¨è½¬æ¥äººå·¥å®¢æœï¼Œè¯·ç¨ç­‰..."
    
    print(f"âš™ï¸  å·¥å…·æ‰§è¡Œç»“æœ: {tool_result[:50]}...")
    
    return {
        **state,
        "tool_result": tool_result,
        "iteration": state.get("iteration", 0) + 1
    }


def knowledge_retriever(state: AgentState) -> AgentState:
    """èŠ‚ç‚¹4ï¼šçŸ¥è¯†åº“æ£€ç´¢"""
    user_query = state["user_query"]
    
    # åœ¨å®é™…åº”ç”¨ä¸­ï¼Œè¿™é‡Œä¼šï¼š
    # 1. å°†é—®é¢˜è½¬ä¸º embedding
    # 2. åœ¨å‘é‡æ•°æ®åº“ä¸­æ£€ç´¢
    # 3. è¿”å›æœ€ç›¸å…³çš„æ–‡æ¡£
    
    # ç®€åŒ–ç‰ˆï¼šä½¿ç”¨ä¹‹å‰çš„çŸ¥è¯†åº“
    kb = SimpleKnowledgeBase()
    results = kb.search(user_query)
    
    if results:
        kb_result = results[0]["answer"]
    else:
        kb_result = "æœªæ‰¾åˆ°ç›¸å…³çŸ¥è¯†"
    
    print(f"ğŸ“š çŸ¥è¯†åº“æ£€ç´¢: {kb_result[:50]}...")
    
    return {
        **state,
        "tool_result": kb_result
    }


def answer_generator(state: AgentState) -> AgentState:
    """èŠ‚ç‚¹5ï¼šç”Ÿæˆå›ç­”"""
    user_query = state["user_query"]
    tool_result = state.get("tool_result", "")
    intent = state["intent"]
    
    # åœ¨å®é™…åº”ç”¨ä¸­ï¼Œè¿™é‡Œä¼šè°ƒç”¨ LLM ç”Ÿæˆè‡ªç„¶è¯­è¨€å›ç­”
    # prompt: "æ ¹æ®ç”¨æˆ·é—®é¢˜ {user_query} å’Œæ£€ç´¢ç»“æœ {tool_result}ï¼Œç”Ÿæˆå›ç­”"
    
    # ç®€åŒ–ç‰ˆï¼šç›´æ¥è¿”å›å·¥å…·ç»“æœ
    final_answer = tool_result if tool_result else "æŠ±æ­‰ï¼Œæˆ‘æ— æ³•å›ç­”è¿™ä¸ªé—®é¢˜"
    
    print(f"ğŸ’¬ ç”Ÿæˆå›ç­”: {final_answer[:50]}...")
    
    return {
        **state,
        "final_answer": final_answer
    }


# ========== 4. è·¯ç”±å‡½æ•° ==========

def should_use_tool(state: AgentState) -> str:
    """å†³å®šæ˜¯å¦ä½¿ç”¨å·¥å…·"""
    if state["need_tool"]:
        return "tool_executor"
    else:
        return "knowledge_retriever"


def should_continue(state: AgentState) -> str:
    """å†³å®šæ˜¯å¦ç»§ç»­è¿­ä»£"""
    iteration = state.get("iteration", 0)
    
    # å¦‚æœè¿­ä»£æ¬¡æ•°è¶…è¿‡3æ¬¡ï¼Œç›´æ¥ç»“æŸ
    if iteration >= 3:
        return "answer_generator"
    
    # å¦‚æœå·²æœ‰å·¥å…·ç»“æœï¼Œç”Ÿæˆå›ç­”
    if state.get("tool_result"):
        return "answer_generator"
    
    # å¦åˆ™ç»§ç»­
    return "tool_executor"


# ========== 5. æ„å»º LangGraph ==========

def create_customer_service_graph():
    """åˆ›å»ºå®¢æœç³»ç»Ÿçš„ LangGraph"""
    
    workflow = StateGraph(AgentState)
    
    # æ·»åŠ èŠ‚ç‚¹
    workflow.add_node("intent_classifier", intent_classifier)
    workflow.add_node("tool_selector", tool_selector)
    workflow.add_node("tool_executor", tool_executor)
    workflow.add_node("knowledge_retriever", knowledge_retriever)
    workflow.add_node("answer_generator", answer_generator)
    
    # è®¾ç½®å…¥å£
    workflow.set_entry_point("intent_classifier")
    
    # æ·»åŠ è¾¹
    workflow.add_edge("intent_classifier", "tool_selector")
    
    # æ¡ä»¶è¾¹ï¼šæ ¹æ®æ˜¯å¦éœ€è¦å·¥å…·é€‰æ‹©è·¯å¾„
    workflow.add_conditional_edges(
        "tool_selector",
        should_use_tool,
        {
            "tool_executor": "tool_executor",
            "knowledge_retriever": "knowledge_retriever"
        }
    )
    
    # ä¸¤æ¡è·¯å¾„éƒ½é€šå‘å›ç­”ç”Ÿæˆ
    workflow.add_edge("tool_executor", "answer_generator")
    workflow.add_edge("knowledge_retriever", "answer_generator")
    
    # ç»“æŸ
    workflow.add_edge("answer_generator", END)
    
    return workflow.compile()


# ========== 6. æµ‹è¯•è¿è¡Œ ==========

print("\n" + "=" * 60)
print("ğŸš€ ç”Ÿäº§çº§ Agentic RAG æ™ºèƒ½å®¢æœï¼ˆLangGraph ç‰ˆï¼‰")
print("=" * 60)

# åˆ›å»º Agent
agent_graph = create_customer_service_graph()

# æµ‹è¯•å¯¹è¯
test_queries = [
    "æŸ¥è¯¢è®¢å• ORDER123456",
    "iPhone 15 çš„ä»·æ ¼å’Œåº“å­˜",
    "å¦‚ä½•é€€è´§ï¼Ÿ"
]

for i, query in enumerate(test_queries, 1):
    print(f"\n{'='*60}")
    print(f"ğŸ’¬ æµ‹è¯• {i}: {query}")
    print(f"{'='*60}\n")
    
    # åˆå§‹çŠ¶æ€
    initial_state = {
        "messages": [],
        "user_query": query,
        "intent": "",
        "need_tool": False,
        "tool_name": "",
        "tool_result": "",
        "final_answer": "",
        "iteration": 0
    }
    
    # è¿è¡Œ Agent
    result = agent_graph.invoke(initial_state)
    
    print(f"\nğŸ¤– æœ€ç»ˆå›ç­”: {result['final_answer']}")
    print(f"ğŸ“Š è¿­ä»£æ¬¡æ•°: {result['iteration']}")


print("\n" + "=" * 60)
print("âœ… LangGraph ç‰ˆæœ¬æ¼”ç¤ºå®Œæˆ")
print("=" * 60)
````
